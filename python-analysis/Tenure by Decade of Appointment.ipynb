```python
%pylab inline
import pandas as pd
from lifelines.utils import datetimes_to_durations
from lifelines import KaplanMeierFitter
```

```python
df = pd.read_csv('../data/parl_data.csv')
df['start_date'] = pd.to_datetime(df['start_date'])
df['end_date'] = pd.to_datetime(df['end_date'])
df['decade'] = df['start_date'].map( lambda d: str(d.year)[:3])
T, C = datetimes_to_durations(df['start_date'], df['end_date'])
df['T'] = T/365.
df['C'] = C
```

```python
kmf = KaplanMeierFitter()
figsize(14,6)

font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 12}

matplotlib.rc('font', **font)


ax = subplot(111)
for decade in sorted(df['decade'].unique(), reverse=True):
    if decade == '201':
        continue
    ix = df['decade'] == decade
    kmf.fit(df.ix[ix]['T'], df.ix[ix]['C'], label=decade + 'x')
    if decade not in ('200', '199'):
        kmf.plot(ax=ax, c='#777777', ci_show=False, alpha = 0.5)
    else:
        kmf.plot(ax=ax, lw=4, alpha=0.9, ci_show=False)

plt.title('Canadian Senate Durations, by Decade Appointed')
plt.xlabel('Years in Senate')
plt.ylabel('Proportion Active after $t$ Years');
```

```python
ix = (df['decade'] == '200') | (df['decade'] == '199') | (df['decade'] == '201')
```

```python
ax = subplot(111)
kmf.fit(df.ix[ix]['T'], df.ix[ix]['C'], label='Post-1990 Appointments')
kmf.plot(ax=ax, c='#348ABD')

kmf.fit(df.ix[~ix]['T'], df.ix[~ix]['C'], label='Pre-1990 Appointments')
kmf.plot(ax=ax, c='#7A68A6')

plt.title('Comparing Senate Durations Pre-1990 and Post-1990')
plt.xlabel('Years in Senate')
plt.ylabel('Proportion Active after $t$ Years');
```

```python
from lifelines.statistics import logrank_test
```

```python
S, P, R = logrank_test(df.ix[ix]['T'], df.ix[~ix]['T'], df.ix[ix]['C'], df.ix[~ix]['C'])
```
